```python
##회문문자열

n = int(input())

for i in range(n):
    s = input()
    # 대소문자 구별 안 하기 위해
    s = s.upper()
    # 문자열 길이 비교
    size = len(s)
    
    # 양 끝부터 값 비교
    for j in range(size//2):
        if s[j] != s[-1-j]:
            print('#%d NO' %(i+1))
            break
    else:
        print('#%d YES' %(i+1))
```

```python
## 숫자만 추출

s = input()
res = 0
for x in s:
    if x.isdecimal():
        res = res*10+int(x)
print(res)

cnt=0
for i in range(1, res+1):
    if res%i==0:
        cnt +=1
print(cnt)
```

```python
## 카드 역배치

a = list(range(21))

for _ in range(10):
    s, e = map(int, input().split())
    # 반복횟수
    for i in range((e-s+1)//2):
        # 데이터 스왑
        a[s+i], a[e-i] = a[e-i], a[s+i] 
a.pop(0)
for x in a:
    print(x, end=' ')
```

```python
## 두 리스트 합치기(sort함수 사용하지 않고 시간복잡도를 줄이는 방법)

n = int(input())
a = list(map(int, input().split()))
m = int(input())
b = list(map(int, input().split()))
p1=p2=0
c = []

while p1<n and p2<m:
    if a[p1] <=b[p2]:
        c.append(a[p1])
        p1 += 1
    else:
        c.append(b[p2])
        p2 += 1
if p1<n:
    c = c+a[p1:]
if p2<m:
    c = c+b[p2:]

for x in c:
    print(x, end=' ')
```

```python
## 격자판 최대합
n = int(input())
grid = []
for _ in range(n):
    data = list(map(int, input().split()))
    grid.append(data)
#print(grid)

total_sum = []
tmp_diag1 = 0
tmp_diag2 = 0
for i in range(n):
    tmp_row = 0
    tmp_col = 0
    tmp_diag1 += grid[i][i]
    tmp_diag2 += grid[i][(n-1)-i]
    for j in range(n):
        tmp_row += grid[i][j]
        tmp_col += grid[j][i]

    total_sum.append(tmp_row)
    total_sum.append(tmp_col)
total_sum.append(tmp_diag1)
total_sum.append(tmp_diag2)

print(max(total_sum))
```

```python
## 사과나무

n = int(input())
grid = [list(map(int, input().split())) for _ in range(n)]
res = 0
s = e = n//2

for i in range(n):
    for j in range(s, e+1):
        print(s, e+1)
        res +=grid[i][j]
    if i < n//2:
        s -= 1
        e += 1
    else:
        s +=1
        e -=1
print(res)
```

```python
## 모래시계 및 값 이동

n = int(input())
grid = [list(map(int, input().split())) for _ in range(n)]

m = int(input())
for i in range(m):
    h, t, k = map(int, input().split())
    if t==0:
        for _ in range(k):
            grid[h-1].append(grid[h-1].pop(0))
    else :
        for _ in range(k):
            grid[h-1].insert(0, grid[h-1].pop())

for x in grid:
    print(x)

res = 0
s = 0
e = n-1

for i in range(n):
    for j in range(s, e+1):  #e+1을 해야 e 까지 돈다
        res += grid[i][j]    # res에 모든 값을 누적
    if i < n//2:             # 절반보다 작을 땐 좁혀야함
        s += 1
        e -= 1
    else:
        s -= 1
        e += 1
print(res)
```
